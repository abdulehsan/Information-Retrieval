**Importing necessary Libraries**

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
import re
import math
import numpy as np
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

Defining the preprocessing function where it takes 1 parameter which will be the value of dictionary in this case the verse.

The preprocessing does the following:

*   Converting the verse to lowercase.
*   Creating an empty list which will store each word in the verse.
*   Splitting every word either by space or "-" or emm dash or enn dash.
*   Removing Stop words
*   Lemmatizing each word by noun and by verb.

```python
def preprocessing(doc):
    doc = doc.lower()
    preprocessed_text = []
    for word in re.split(r'\s|-|\u2014|\u2013|,', doc):
        word = re.sub(r'[^\w\s]', '', word)
        if word and word not in stop_words:
            word = lemmatizer.lemmatize(word)
            word = lemmatizer.lemmatize(word, pos='v')
            preprocessed_text.append(word)
    return preprocessed_text
```

*   Passing each verse to the preprocessing function.
*   Creating a dictionary that stores each preprocessed verse with its verse number as key.

```python
preprocessed_verses = {}
for key, verse in verses.items():
    preprocessed_verses[key] = preprocessing(verse)
print(preprocessed_verses)
```

*   Creating a set of vocabulary

```python
vocab = set()
for words in preprocessed_verses.values():
    vocab.update(words)
```

```python
print(vocab)
```

Defining calculating tf log function that will take preprocessed verses dictionary and vocabulary as parameters.
It will count the word in each verse


```python
def cal_tf_log(preprocessed_verses, vocab):
    tf = {}
    for word in vocab:
        tf[word] = []
        for verse_words in preprocessed_verses.values():
            count = verse_words.count(word)
            if count > 0:
              log_tf = 1 + math.log10(count)
            else:
              log_tf = 0
            tf[word].append(log_tf)
    return tf
```

```python
tf_log = cal_tf_log(preprocessed_verses, vocab)
tf_log
```

Defining function for calculating IDF that will take the same parameters as tf log function.
It will count the word in whole document.

```python
def idf(preprocessed_verses, vocab):
    idf = {}
    for word in vocab:
      count = 0
      idf[word] = 0
      for verse_words in preprocessed_verses.values():
        if word in verse_words:
          count+=1
      idf[word]=math.log10(len(preprocessed_verses)/count)
    return idf
```

```python
idf_matrix = idf(preprocessed_verses, vocab)
idf_matrix

```

Defining function for calculating TF-IDF , that will take tf log dictionary and idf dictionary as parameters and will simply multiply them.

```python
def tf_idf(tfLog,idf):
  tf_idf = {}
  for key,value in tfLog.items():
    tf_idf[key] = []
    for value in tfLog[key]:
      tf_idf[key].append(value*idf[key])
  return tf_idf
```

```python
tf_idf_matrix = tf_idf(tf_log,idf_matrix)
tf_idf_matrix
```

Defining a function for Cosine Similarity that will take query vector and document vector as a parameters, and calculate cosine similarity between them.

```python
def cosine_similarity(qvector,docvector):
  # print(len(qvector))
  # print(len(docvector))
  qmag = np.linalg.norm(qvector)
  dmag = np.linalg.norm(docvector)
  dotproduct = np.dot(qvector,docvector)
  return dotproduct/(qmag*dmag)
```

* Passing each document vector & query vector to the cosine similarity function.

Defining function for Precision & Recall , that will take retrieved verses dictionary and ground truth dictionary.

For calculating precision and recall we need the number of actual relevant documents.
    
We can calculate it by taking intersection of retrieved and ground truth.

```python
def cal_precision_recall(retrieved_verses,ground_truth):
  precision = {}
  recall = {}
  actual_retrieved = []
  for query_id,verse_numbers in retrieved_verses.items():
    if len(verse_numbers)>0:
        for verse_number in verse_numbers:
          if verse_number in ground_truth[query_id]:
            actual_retrieved.append(verse_number)
        precision[query_id] = len(actual_retrieved)/len(verse_numbers)
    else:
        precision[query_id] = 0

    if len(ground_truth[query_id])>0:
        recall[query_id] = len(actual_retrieved)/len(ground_truth[query_id])
    else:
        recall[query_id] = 0

    actual_retrieved = []

  return precision,recall
```

Defining function for calculating F1-Score

```python
def cal_f1_score(precision, recall):
  f1_score = {}
  for query_id in precision.keys():
    denominator = precision[query_id] + recall[query_id]
    if denominator == 0:
      f1_score[query_id] = 0
    else:
      f1_score[query_id] = 2 * (precision[query_id] * recall[query_id]) / denominator
  return f1_score
```

```python
f1_score = cal_f1_score(precision,recall)
f1_score
```

**Strengths of the System**

* Simple yet Effective IR Approach: Using TF-IDF + cosine similarity provides a proven and interpretable method for information retrieval.

* Proper Text Preprocessing: The system incorporates standard NLP techniques: stopword removal, lemmatization, and tokenization — all crucial for reducing noise in vector construction.

* Vector Space Model: Representing documents and queries in vector space allows for flexible similarity calculations using cosine distance.

* Manual Evaluation: The inclusion of a ground truth and computation of precision, recall, and F1-score provides a strong foundation for system evaluation and comparison.

**Weaknesses and Limitations**
* Short and Sparse Queries: Most queries are single words. These often lack context, leading to semantic gaps between queries and verses. For example, "Prayer" may not match if the word isn't used exactly in the text.

* No Semantic Understanding: The system only matches exact lemmatized terms. There’s no synonym handling, word embeddings, or deeper semantic matching (like with BERT or Word2Vec), which means it can miss conceptually relevant verses.

* TF-IDF Limitations: TF-IDF assumes term independence, which isn’t always realistic.It gives high weight to rare words, which might not be semantically important.

* Hardcoded Ground Truth: Ground truth is manually written and not scalable. For larger corpora or user-driven systems, automatic labeling or relevance feedback would be necessary.
